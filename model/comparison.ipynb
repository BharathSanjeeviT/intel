```python
INTEL_EXECUTION_TIME = 25 #MINS
INTEL_EXECUTION_TIME_WITH_GPU = 20 #MINS
EXECUTION_TIME = 38 #MINS
```

```python
x = np.array(["EXECUTION_TIME", "INTEL_EXECUTION_TIME"])
x1=np.array(["EXECUTION_TIME", "INTEL_EXECUTION_TIME_WITH_GPU"])
y = np.array([EXECUTION_TIME, INTEL_EXECUTION_TIME])
y1 =np.array([EXECUTION_TIME, INTEL_EXECUTION_TIME_WITH_GPU])
plt.bar(x,y,color=["orange","blue"])
plt.show()
```

```python
plt.bar(x1,y1,color=["orange","blue"])
plt.show()
```
